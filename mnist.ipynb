```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from layer import Layer
from neural_network import NeuralNetwork
```

```python
d = './MNIST_CSV/'
train = pd.read_csv(d + 'mnist_train.csv', header=None)
test = pd.read_csv(d + 'mnist_test.csv', header=None)

train = np.array(train)
test = np.array(test)

np.random.shuffle(train)
np.random.shuffle(test)
```

```python
train = train.T
test = test.T

train_X = train[1:] / 255
train_Y = train[0]

test_X = test[1:] / 255
test_Y = test[0]
```

```python
train_X.shape, test_X.shape
```

```python
l1 = Layer(10, 784, 'relu')
l2 = Layer(10, 10, 'softmax')

nn = NeuralNetwork()
nn.add_layer(l1)
nn.add_layer(l2)

nn.gradient_descent(train_X, train_Y, -3, 0.15)
```

```python
def see_prediction(index, nn, x, y):
    current_image = x[:, index, None]
    prediction = nn.predict(x[:, index, None])
    label = y[index]
    print("Prediction: ", prediction)
    print("Label: ", label)
    
    current_image = current_image.reshape((28, 28)) * 255
    plt.gray()
    plt.imshow(current_image, interpolation='nearest')
    plt.show()
```

```python
see_prediction(5, nn, test_X, test_Y)
see_prediction(6, nn, test_X, test_Y)
```
